 **WHAT IS ENCODING**

*   Encoding is the process of converting data from one format to another. In the context of data preprocessing for machine learning, encoding typically refers to transforming categorical data (data that can be divided into different categories) into a numerical format that can be used by machine learning algorithms. Most machine learning algorithms require numerical input, so encoding is a crucial step in data preprocessing.





**ENODING TECHNIQUES**

*Label encoding*



*  Each unique category value is assigned a numerical value.
*   Label encoding is a technique used in machine learning and data analysis to convert categorical variables into numerical format.




```python
import pandas as pd
```

```python
df=pd.read_csv('/student.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['class']=le.fit_transform(df['class'])
```

Here in the output class is encoded as 1=TE and 0=BE and the name is encoded by 0,1,2,3 etc by alphabet order

```python
df
```

Here age is encoded

```python
df['age']=le.fit_transform(df['age'])
df
```

 *One Hot Encodin*


*   Converts each category value into a new binary column (0 or 1).
*    Suitable for nominal data where the order does not matter.



When you apply one-hot encoding to the class column, it will create new binary columns for each unique category (i.e., TE and BE).

```python
df
```

```python
df1= pd.get_dummies(df, columns=['class'])
df1
```
