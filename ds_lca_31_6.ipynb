**SCATTER PLOT**


```python

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd

iris = load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target_names[y]

sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species', palette='Set1')

plt.title('Iris Dataset - Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

plt.show()
```

**HISTOGRAM**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd
iris = load_iris()
X = iris.data
y = iris.target
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target_names[y]

plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='sepal length (cm)', hue='species', palette='Set1')

plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

**LINE CHART**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd

iris = load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target_names[y]

avg_df = df.groupby('species').mean()

plt.figure(figsize=(8, 6))
sns.lineplot(data=avg_df, markers='o', dashes=False, palette='Set1')

plt.title('Average Sepal Length and Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Average Measurement (cm)')
plt.show()
```

**BOX PLOT**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd

iris = load_iris()
X = iris.data
y = iris.target

df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target_names[y]

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='sepal length (cm)', hue='species', palette='Set1')

plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')

plt.show()
```
