**DATA VISUALIZATION**

**MATPLOTLIB**



*   Matplotlib is a popular plotting library for the Python programming language. It provides a flexible and comprehensive framework for creating a wide range of static, animated, and interactive visualizations in Python. It is widely used for data visualization


**LINE PLOT**



*   A line plot is useful for visualizing trends over time or a continuous variable.



```python
df=pd.read_csv('Salary_Data.csv')
```

```python
df.head()
```

```python
xpoints = df['YearsExperience']
ypoints = df['Salary']
plt.plot(xpoints, ypoints)
plt.xlabel('Years experience')
plt.ylabel('Salary')
plt.title('Line plot')
```

**LINE MARKERS**


*   In Matplotlib, you can add markers to your lines to highlight individual data points. The plt.plot() function allows you to specify markers along with line styles and colors.

*   Common Marker Styles:

.: Point marker

o: Circle marker

v: Triangle down marker

^: Triangle up marker

s: Square marker

p: Pentagon marker

*: Star marker

+: Plus marker

x: Cross marker

D: Diamond marker

h: Hexagon marker






```python
xpoints = df['YearsExperience']
ypoints = df['Salary']
plt.plot(xpoints, ypoints)
plt.xlabel('Years experience')
plt.ylabel('Salary')
plt.title('Line plot')

plt.plot(xpoints, ypoints, marker='o', linestyle='-', color='hotpink')
```

**STYLING LINES**


*   Matplotlib offers several predefined line styles:

'-': Solid line (default)

'--': Dashed line

'-.': Dash-dot line

':': Dotted line




```python
xpoints = df['YearsExperience']
ypoints = df['Salary']
plt.xlabel('Years experience')
plt.ylabel('Salary')
plt.title('Line plot')
plt.plot(xpoints, ypoints * 0.5, linestyle='--', color='hotpink', label='Dashed Line')
```

**COUNT PLOT**


*    Count plot is a type of bar plot that displays the frequency of each category in a categorical variable. It's particularly useful when you want to quickly visualize the distribution of categorical data.



```python
df1=pd.read_csv('Iris.csv')
df1.head()
```

```python
count_data = df1['Species'].value_counts()
count_data.plot(kind='bar', color='skyblue')
plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
```

**PIE CHART**


*   A pie chart is a great way to visualize the proportion of each category in your data.



```python
count_data = df1['Species'].value_counts()
plt.figure(figsize=(8, 8))
count_data.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'salmon'])
plt.title('Proportion of Each Iris Species')
```

**HORIZONTAL COUNTPLOT**

```python
count_data = df1['Species'].value_counts()
plt.figure(figsize=(10, 6))
count_data.plot(kind='barh', color='skyblue')
plt.title('Count of Each Iris Species')
plt.xlabel('Count')
plt.ylabel('Species')
```
